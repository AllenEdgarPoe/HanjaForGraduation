## 졸업을 위한 한자 시험 테스트

```python
#data detecting
import re, numpy as np, random, time
with open('hanja_rawdata.txt',encoding = 'utf-8') as test:
    texts = test.read().split('\n')
    
problem_1 = []    #chinese = meaning+sound
problem_2 = []    #chinese = sound
for text in texts:
    b = re.search('(.+)?<(.+)?>',text).groups()
    if ' ' in b[1]:
        text = []
        chinese = b[0]
        meaning = b[1].split(' ')[0]
        sound = b[1].split(' ')[1]
        text.append(chinese)
        text.append(meaning)
        text.append(sound)
        problem_1.append(text)
    else:
        text = []
        chinese = b[0]
        sound = b[1]
        text.append(chinese)
        text.append(sound)
        problem_2.append(text)
        
#problem_1
chinese = np.array([a[0] for a in problem_1])
meaning = np.array([a[1] for a in problem_1])
sound = np.array([a[2] for a in problem_1])

print('''
    본 시험은 고려대학교 졸업을 위한 한자 시험 대비를 위한 테스트 입니다.  
    총 4개의 파트로 구분되어있습니다.
    각 파트는 한자 의미 고르기, 한자 독음 고르기, 두 글자 이상의 독음 고르기, 한글의 알맞은 한자 표기 고르기 입니다.
    ''')

time.sleep(1)
num_questions = int(input('한자 모의 시험을 시작합니다. test(원하는 문제수)를 입력하십시오 '))

def test(number):
    def meaning_test():
        print('파트1. 알맞은한자의 의미를 고르시오')
        a = np.random.randint(0,len(chinese),number)
        answer = meaning[a]
        for i in a:
            bogi = list(np.r_[np.random.randint(0,len(meaning),4),np.array([i])])
            random.shuffle(bogi)
            print(chinese[i],'\t',meaning[bogi])
        t = input('답을 입력하시오: ')
        return('파트1 답: {}'.format(list(answer)))

    def sound_test():
        print('\n파트2. 알맞은 한자의 독음을 고르시오')
        a = np.random.randint(0,len(chinese),number)
        answer = sound[a]
        for i in a:
            bogi = list(np.r_[np.random.randint(0,len(meaning),4),np.array([i])])
            random.shuffle(bogi)
            print(chinese[i],'\t',sound[bogi])
        t = input('답을 입력하시오: ')
        return('파트2 답: {}'.format(list(answer)))   

    #problem_2
    chinese2 = np.array([i[0] for i in problem_2])
    sound2 = np.array([i[1] for i in problem_2])
    def character_to_sound_test():
        print('\n파트3. 각 단어의 알맞은 의미를 고르시오')
        a = np.random.randint(0,len(chinese2),number)
        answer = sound2[a]
        for i in a:
            new = np.array([j for j in sound2 if len(j)==len(sound2[i])])
            bogi = np.append(new[np.random.randint(0,len(new),4)],sound2[i])
            t = np.random.shuffle(bogi)
            print(chinese2[i],'\t',bogi[t])
        t = input('답을 입력하시오: ')
        return('파트3 답: {}'.format(list(answer)))

    def sound_to_character_test():
        print('\n파트4. 각 한글의 알맞은 한자 독음을 고르시오')
        a = np.random.randint(0,len(chinese2),number)
        answer = chinese2[a]
        for i in a:
            new = np.array([j for j in chinese2 if len(j)==len(chinese2[i])])
            bogi =np.append(new[np.random.randint(0,len(new),4)],chinese2[i])
            t = np.random.shuffle(bogi)
            print(sound2[i],'\t',bogi[t])
        t = input('답을 입력하시오: ')
        return('파트4 답: {}'.format(list(answer)))
            
    return meaning_test(),sound_test(), character_to_sound_test(), sound_to_character_test()

test(num_questions)
```
